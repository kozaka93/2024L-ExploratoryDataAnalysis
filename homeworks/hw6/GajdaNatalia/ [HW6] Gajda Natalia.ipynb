# Analiza zmiennych w zależności od przynależności do grupy (należy wykorzystać GroupBy względem zmiennej jakościowej)

```python
df2 = df.groupby("Type_of_Renewable_Energy")["Initial_Investment_USD"].agg(['sum', 'mean']).reset_index()
df2
```

Okazuje się, że dla każdego typu energii odnawialnej początkowe inwestycje w nią były dosyć podobne (ten sam rząd). Jednakże, patrząc na średnią inwestycji możemy wnioskować, że w niektóre inwestycje w konkretnym typie energii włożono zdecyowanie mniej pieniędzy niż w inne. 

```python
df2 = df.groupby("Type_of_Renewable_Energy")[["Installed_Capacity_MW", "Energy_Production_MWh", "Energy_Consumption_MWh","Energy_Storage_Capacity_MWh","Initial_Investment_USD", "Financial_Incentives_USD", "GHG_Emission_Reduction_tCO2e", "Air_Pollution_Reduction_Index","Jobs_Created"]].mean().reset_index()
df2
```

Średnia wartości konkretnych zmiennych ze względu na typ energii odnawialnej (użycie groupby) :) 

# Wykresy analizujące rozkład zmiennych w zbiorze danych razem z wnioskami

```python
sns.boxplot(data=df[["Energy_Production_MWh", "Energy_Consumption_MWh"]], color='w')
plt.show()
```

Powyżej boxplot pokazujący rozkład produkcji i konsumpcji energii.
Oba bardzo przypominają rozkład normalny. Rozkład sugeruje również bardzo szeroki zakres wartości reprezentujących zużycie i konsumpcję energii. 


```python
sns.violinplot(data=df["Initial_Investment_USD"], color = 'w')
plt.show()
```

Rozkład wygląda na symetryczny. Widać też, że w danych nie występują raczej outliery. Wartości bliskich medianie jest bardzo dużo. 

# Wykresy wskazujące zależności w zbiorze danych z wnioskami

```python
df1 = df.groupby("Type_of_Renewable_Energy")["Air_Pollution_Reduction_Index"].agg(['sum', 'mean']).reset_index()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.bar(df1["Type_of_Renewable_Energy"], df1["mean"])
plt.show()
df1
```

Wykres przedstawia średnią wartość wskaźnika redukcji zanieczyszczenia powietrza dla każdego typu odnawialnej energii. Widać, że bez względu na typ, jest on praktycznie taki sam. Na wykresie bardzo słabo to widać, ale wywołana tabelka przedstawia dokładne wartości, zatem możemy powiedzieć, że najlepszy wskaźnik mamy dla typu 1, najgorszy dla 2.  

```python
g = sns.catplot(x='Type_of_Renewable_Energy', 
                   y='GHG_Emission_Reduction_tCO2e', 
                   data=df, 
                   hue='Funding_Sources',  
                   col='Funding_Sources',  
                   kind='box') 
  
plt.show()
```

Mamy podobną sytuację jak poprzednio. Tym razem mamy 3 boxploty, każdy dla innego źródła dofinansowania, a w każdym boxplocie podział na typy energii odnawialnej. Wnioski są takie, że bez względu na źródło dofinansowania oraz typ energii rozkłady redukcji gazów cieplarnianych są praktycznie takie same, co oznaczałoby, że są one bardzo podobnie skuteczne i ciężko wybrać najlepszy.
